```python
# Distribution of Variable Amount
plt.figure(figsize=(10, 6))
sns.histplot(credit_card_data['Amount'], bins=50, kde=True)
plt.title('Distribution des montants des transactions')
plt.xlabel('Montant')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Distribution of Variable Time
plt.figure(figsize=(10, 6))
sns.histplot(credit_card_data['Time'], bins=50, kde=True)
plt.title('Distribution des moments des transactions')
plt.xlabel('Temps (secondes)')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Distribution of 'Amount' for legitim and fraud transactions
plt.figure(figsize=(10, 6))
sns.histplot(credit_card_data[credit_card_data['Class'] == 0]['Amount'], color='blue', bins=50, kde=True, label='Légitimes')
sns.histplot(credit_card_data[credit_card_data['Class'] == 1]['Amount'], color='red', bins=50, kde=True, label='Frauduleuses')
plt.title('Distribution des montants pour les transactions légitimes et frauduleuses')
plt.xlabel('Montant')
plt.ylabel('Fréquence')
plt.legend()
plt.show()
```
